```python
import pandas as pd
import numpy as np
```

```python
# import data
data_dir = r"C:\Users\gcmar\Dropbox (UFL)\sleap_umap_tube_test_03312023\tube_test_analysis\all_matches_videos"
df_coord = pd.read_csv(data_dir+"\\"+"all_matches_videos_node_coordinates.csv", index_col = "Unnamed: 0")
df_coord = df_coord.reset_index(drop=True)
df_dist = pd.read_csv(data_dir+"\\"+"all_matches_videos_node_distances.csv", index_col = "Unnamed: 0")
df_angle = pd.read_csv(data_dir+"\\"+"all_matches_videos_node_angles.csv", index_col = "Unnamed: 0")

# add all spatial data together
df_spatial = pd.DataFrame()
df_spatial['trial'] = df_coord['trial']
df_spatial = pd.concat([df_spatial, df_coord.drop('trial', axis=1), df_dist.drop('trial', axis=1), df_angle.drop('trial', axis=1) ], axis=1)
df_spatial.to_csv(data_dir+"\\"+"all_matches_videos_spatial.csv") # save the data to disk
```

```python
# import data anew
df_spatial = pd.read_csv(data_dir+"\\"+"all_matches_videos_spatial.csv", index_col = "Unnamed: 0")
```

```python
# add temporal component
df_temporal = pd.DataFrame()
df_temporal['trial'] = df_spatial['trial']
df_temp = df_spatial.drop('trial', axis=1).diff()
df_temporal = pd.concat([df_temporal, df_temp ], axis=1)
# remove Nan values and replace with value that comes after
df_temporal.iloc[0] = df_temporal.iloc[1]

# save the temporal data to disk
df_temporal.to_csv(data_dir+"\\"+"all_matches_videos_temporal.csv") # save the data to disk
```
